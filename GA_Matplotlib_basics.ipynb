```python
import matplotlib.pyplot as plt
```

```python
X= [1,4,6,7,8,20,12,23]
y= [0,3,5,5,8,10,14,15]
```

```python
plt.plot(X,y)
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Draw a line')
plt.show()
```

```python
x1 = [10,20,30]
y1 = [20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Multiple line plots')
plt.plot(x1,y1,label="Line 1-w3",color='blue', linewidth=3)
plt.plot(x2,y2,label="Line 2-w3",color='red',linewidth=3)
plt.legend()
```
