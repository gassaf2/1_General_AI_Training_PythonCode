```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python

#iris.species.unique()
#iris['species'].unique()
iris['species'].value_counts()
```

```python
sns.set(style="whitegrid")
sns.relplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, size=5)
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
```

```python
sns.catplot(x='species', y='petal_length', data=iris, kind="box")
```

```python
sns.catplot(x='species', y='petal_length', kind='violin', data=iris)
```

```python
sns.set(style='white')
sns.pairplot(iris,hue="species",height=2) # You can even add a hue here!
plt.show()
```
